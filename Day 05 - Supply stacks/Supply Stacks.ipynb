## Part 1
The expedition can depart as soon as the final supplies have been unloaded from the ships. Supplies are stored in stacks of marked crates, but because the needed supplies are buried under many other crates, the crates need to be rearranged.

The ship has a giant cargo crane capable of moving crates between stacks. To ensure none of the crates get crushed or fall over, the crane operator will rearrange them in a series of carefully-planned steps. After the crates are rearranged, the desired crates will be at the top of each stack.

The Elves don't want to interrupt the crane operator during this delicate procedure, but they forgot to ask her which crate will end up where, and they want to be ready to unload them as soon as possible so they can embark.

They do, however, have a drawing of the starting stacks of crates and the rearrangement procedure (your puzzle input). For example:

```
    [D]   
[N] [C]    
[Z] [M] [P]
 1   2   3 
 ```

move 1 from 2 to 1
move 3 from 1 to 3
move 2 from 2 to 1
move 1 from 1 to 2
In this example, there are three stacks of crates. Stack 1 contains two crates: crate Z is on the bottom, and crate N is on top. Stack 2 contains three crates; from bottom to top, they are crates M, C, and D. Finally, stack 3 contains a single crate, P.

Then, the rearrangement procedure is given. In each step of the procedure, a quantity of crates is moved from one stack to a different stack. In the first step of the above rearrangement procedure, one crate is moved from stack 2 to stack 1, resulting in this configuration:

```
[D]        
[N] [C]    
[Z] [M] [P]
 1   2   3 
 ```
 
In the second step, three crates are moved from stack 1 to stack 3. Crates are moved one at a time, so the first crate to be moved (D) ends up below the second and third crates:

```
        [Z]
        [N]
    [C] [D]
    [M] [P]
 1   2   3
 ```
 
Then, both crates are moved from stack 2 to stack 1. Again, because crates are moved one at a time, crate C ends up below crate M:

```
        [Z]
        [N]
[M]     [D]
[C]     [P]
 1   2   3
 ```
 
Finally, one crate is moved from stack 1 to stack 2:

```
        [Z]
        [N]
        [D]
[C] [M] [P]
 1   2   3
 ```
 
The Elves just need to know which crate will end up on top of each stack; in this example, the top crates are C in stack 1, M in stack 2, and Z in stack 3, so you should combine these together and give the Elves the message CMZ.

After the rearrangement procedure completes, what crate ends up on top of each stack?

````python
def parse_stacks(stack_lines, num):
    """Take the lines of the text file describing #num stacks, return a 2D list of the stacks.

        Args:
            stack_lines (list): Each element of the list is a line of the text file.
            num (int): Number of stacks.
            
        Returns: 
            stacks (list): 2D list of the stacks.
        """
    stacks = [ []*3 for i in range(num)]
    for index in range(len(stack_lines)):
        current_line = stack_lines[-1 * (index + 1)]
        for stack_index in range(num):
            position_index = 4 * stack_index + 1
            if len(current_line) >= position_index + 1:
                current_character = current_line[position_index]
                if ord(current_character) >= 65 and ord(current_character) <= 90:
                    stacks[stack_index].append(current_character)
    return stacks


def move(stacks, instruction):
    """Applies an instruction to a stack state.

    Args:
        stacks (list): The stack state.
        instruction (list): An instruction, e.g. [1, 2, 1] for "move 1 from 2 to 1".

    Returns:
        None (changes the object passed to it).
    """
    num_to_move = instruction[0]
    from_stack_index = instruction[1] - 1
    to_stack_index = instruction[2] - 1
    for index in range(num_to_move):
        stacks[to_stack_index] += stacks[from_stack_index][-1:]
        del stacks[from_stack_index][-1:]
        

def get_tops(stacks):
    """Finds the letter at the top of each stack, concatenates into a string.

    Args:
        stacks (list): The stack state.

    Returns:
        The string of top letters.
    """
    output = ""
    for stack in stacks:
        output += stack[-1]
    return output

def convert_instruction_line_to_list(line):
    """Takes a text instruction and returns it formatted as a list.

    Args:
        line: The line of text, e.g. "move 1 from 2 to 1".

    Returns:
        A list representing the instruction, e.g. [1, 2, 2].
    """
    line_split = line.split(" ")
    return int(line_split[1]), int(line_split[3]), int(line_split[5])


with open("input.txt") as f:
    lines = f.read().splitlines()
    
# Find the position of the stack numbers.
found_stack_numbers = False
current_line = 0
while not found_stack_numbers:
    if lines[current_line][1] == "1":
        stack_numbers_position = current_line
        found_stack_numbers = True
    current_line += 1

# Determine the number of stacks
number_of_stacks = int(lines[stack_numbers_position][-2])

# Split the lines up into the stack lines and the instruction lines.
stack_lines = lines[:stack_numbers_position]
instruction_lines = lines[stack_numbers_position + 2:]

# Convert the instructions to lists.
instructions_lists = []
for line in instruction_lines:
    instructions_lists.append(convert_instruction_line_to_list(line))

# Parse the stacks.
stacks = parse_stacks(stack_lines, number_of_stacks)

# Apply the instructions.
for num, instruction in enumerate(instructions_lists):
    move(stacks, instruction)

# Work out answer and print.
print(get_tops(stacks))
````
